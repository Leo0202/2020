```python
import pandas as pd
from geopy.geocoders import Nominatim
import time


def get_address(lat, lon, geolocator):
    try:
        location = geolocator.reverse((lat, lon), exactly_one=True)
        return location.address if location else "No encontrado"
    except:
        return "Error al convertir"

def process_file(file_path, output_file_path):
    # Cargar el archivo Excel
    df = pd.read_excel(file_path, sheet_name='Hoja2', engine='openpyxl')

    # Inicializar el geocodificador
    geolocator = Nominatim(user_agent="geoapiExercises")

    # Procesar cada fila del archivo
    for index, row in df.iterrows():
        try:
            # Convertir coordenadas a direcciones
            address_a = get_address(row['Lat. Inicial'], row['Lng. Inicial'], geolocator)
            address_b = get_address(row['Lat. Final'], row['Lng. Final'], geolocator)

            # Guardar resultados en el DataFrame
            df.at[index, 'DIRECCION DEL PUNTO A (INICIAL)'] = address_a
            df.at[index, 'DIRECCION DEL PUNTO B (FINAL)'] = address_b

            # Pausar entre solicitudes
            time.sleep(1)
        except Exception as e:
            print(f"Error procesando la fila {index}: {e}")

    # Guardar el resultado en un nuevo archivo Excel
    df.to_excel(output_file_path, index=False, sheet_name='Hoja2')
    print(f"Proceso completado. Archivo guardado como: {output_file_path}")


if __name__ == "__main__":
    # Ruta del archivo de entrada
    input_file = "https://raw.githubusercontent.com/Leo0202/Leonardo-Mora/master/PRUEBA2.xlsx"  # Cambia esta ruta por la de tu archivo
    output_file = "./PARACHATGPT_Processed.xlsx"

    process_file(input_file, output_file)

    # codigo inicial para convertir coordenadas
```
